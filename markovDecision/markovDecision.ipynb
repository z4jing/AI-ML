```
import texttable as tt
import copy



inputsquare = [['','1','2','3','4'],
               ['1',0,0,0,0],
               ['2',0,'X',0,-1],
               ['3',0,0,0,1]]

def myprint(allsquare):
  table = tt.Texttable()
  for row in allsquare:
    table.add_row(row)
  print(table.draw(),"\n")
  

############################
def hitedge(pair,rowl,coll):
  a = pair[0]
  b = pair[1]
  if a < 1:
    return (a+1,b)
  elif a >= coll:
    return (a-1,b)
  elif b < 1:
    return (a,b+1)
  elif b >= rowl:
    return (a,b-1)
  else:
    return (a,b)
    
def hitwall(pair,di):
  a = pair[0]
  b = pair[1]
  if a == 2 and b == 2:
    if di == "up":
      return (a+1,b)
    elif di == "down":
      return (a-1,b)
    elif di == "left":
      return (a,b+1)
    elif di == "right":
      return (a,b-1)
  else:
    return (a,b)

def UPDATE(allsquare,R,gamma):
  rowl = len(allsquare[0])
  coll = len(allsquare)
  tem = copy.deepcopy(allsquare)
  for a in range(1,coll):
    for b in range(1,rowl):
      if (a != 2 or b != 2) and (a != 2 or b != 4) and (a != 3 or b != 4):
        uptem = hitwall(hitedge((a-1,b),rowl,coll),"up")
        downtem = hitwall(hitedge((a+1,b),rowl,coll),"down")
        lefttem = hitwall(hitedge((a,b-1),rowl,coll),"left")
        righttem = hitwall(hitedge((a,b+1),rowl,coll),"right")
        up = 0.8 * allsquare[uptem[0]][uptem[1]] + 0.1 * allsquare[lefttem[0]][lefttem[1]] + 0.1 * allsquare[righttem[0]][righttem[1]]
        down = 0.8 * allsquare[downtem[0]][downtem[1]] + 0.1 * allsquare[lefttem[0]][lefttem[1]] + 0.1 * allsquare[righttem[0]][righttem[1]]
        left = 0.8 * allsquare[lefttem[0]][lefttem[1]] + 0.1 * allsquare[uptem[0]][uptem[1]] + 0.1 * allsquare[downtem[0]][downtem[1]]
        right = 0.8 * allsquare[righttem[0]][righttem[1]] + 0.1 * allsquare[uptem[0]][uptem[1]] + 0.1 * allsquare[downtem[0]][downtem[1]]
        four = [up,down,right,left]
        tem[a][b] = R + gamma * max(four)
  return tem

def VIA(allsquare,R,gamma):
  iter = 0
  print("Input is:")
  myprint(allsquare)
  rowl = len(allsquare[0])
  coll = len(allsquare)
  count = 0
  while count != 12:
    iter += 1
    count = 0
    tem = UPDATE(allsquare,R,gamma)
    for a in range(1,coll):
      for b in range(1,rowl):
        if tem[a][b] == allsquare[a][b]:
          count += 1
        else:
          if round(tem[a][b],3) == round(allsquare[a][b],3):
            count += 1
    allsquare = tem
    print("State "+str(iter)+" is:")
    myprint(allsquare)
  return allsquare

def MDP(allsquare):
  rowl = len(allsquare[0])
  coll = len(allsquare)
  tem = copy.deepcopy(allsquare)
  for a in range(1,coll):
    for b in range(1,rowl):
      if (a != 2 or b != 2) and (a != 2 or b != 4) and (a != 3 or b != 4):
        uptem = hitwall(hitedge((a-1,b),rowl,coll),"up")
        downtem = hitwall(hitedge((a+1,b),rowl,coll),"down")
        lefttem = hitwall(hitedge((a,b-1),rowl,coll),"left")
        righttem = hitwall(hitedge((a,b+1),rowl,coll),"right")
        up = 0.8 * allsquare[uptem[0]][uptem[1]] + 0.1 * allsquare[lefttem[0]][lefttem[1]] + 0.1 * allsquare[righttem[0]][righttem[1]]
        down = 0.8 * allsquare[downtem[0]][downtem[1]] + 0.1 * allsquare[lefttem[0]][lefttem[1]] + 0.1 * allsquare[righttem[0]][righttem[1]]
        left = 0.8 * allsquare[lefttem[0]][lefttem[1]] + 0.1 * allsquare[uptem[0]][uptem[1]] + 0.1 * allsquare[downtem[0]][downtem[1]]
        right = 0.8 * allsquare[righttem[0]][righttem[1]] + 0.1 * allsquare[uptem[0]][uptem[1]] + 0.1 * allsquare[downtem[0]][downtem[1]]
        four = [up,down,right,left]
        tem[a][b] = ""
        for each in four:
          if each == max(four):
            if each == up:
              tem[a][b] += "^"
            elif each == down:
              tem[a][b] += "v"
            elif each == left:
              tem[a][b] += "<"
            elif each == right:
              tem[a][b] += ">"
  print("Optimal policy is:")
  return tem

myprint(MDP(VIA(inputsquare,-1,1)))
      
```
